```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1 = np.linspace(0,10,20)
```

```python
data2 = data1 ** 2
```

```python
plt.subplots()
```

```python
my_fig, my_axes = plt.subplots()
```

```python
my_axes.plot(data1, data2, color="#362825", alpha=0.5)
my_axes.plot(data2, data1, color="#E8B4A9")
plt.show()
```

```python
(new_fig, new_axes) = plt.subplots()

new_axes.plot(data1, data1 + 2, color="blue", linewidth=1.0)
new_axes.plot(data1, data1 + 3, color="yellow", linewidth=1.0)

new_axes.plot(data1, data1 + 3, color="#692819", linestyle="-.")
new_axes.plot(data1, data1 + 5, color="#692819", linestyle=":")
new_axes.plot(data1, data1 + 6, color="#692819", linestyle="--")

new_axes.plot(data1, data1 + 7, color="#000000", linestyle="--", marker="o", markersize=10, markerfacecolor="red")
new_axes.plot(data1, data1 + 8, color="#000000", linestyle="--", marker="+",markersize=5)

plt.show()
```
