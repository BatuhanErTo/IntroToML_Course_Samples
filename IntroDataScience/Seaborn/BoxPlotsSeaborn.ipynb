```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.array([5, 7, 9, 15, 20, 22, 25, 30, 32, 35, 37, 40, 50, 55, 60, 100])

plt.figure(figsize=(6, 5))
sns.boxplot(y=data)
plt.title("Box Plot of Simple Numeric Data")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
data = np.array([5, 7, 9, 15, 20, 22, 25, 30, 32, 35, 37, 40, 50, 55, 60,100,500])

plt.figure(figsize=(6, 5))
sns.boxplot(y=data)
plt.title("Box Plot of Simple Numeric Data")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```
