```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("athlete_events.csv")
```

```python
data.describe()
```

```python
data.isnull()
```

```python
data.columns
```

```python
data.info()
```

```python
plt.scatter("Height", "Weight", data=data)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(data=data, x="Height", y="Weight")
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(data=data, x="Height", y="Weight")
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(data=data, x="Height", y="Weight", hue="Sex")
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(data=data, x="Height", y="Weight", hue="Sex", style="Medal")
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(data=data, x="Height", y="Weight", hue="Sex", style="Medal", size="Medal")
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lineplot(data=data, x="Height", y="Weight", hue="Sex")
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.displot(data=data, x="Height", hue="Sex")
plt.ylabel("Frequency")
plt.title("Atheletes Height Histogram")
plt.show()
```

```python
#kde - kernel density estimate
```

```python
sns.set_style("whitegrid")
sns.displot(data=data, x="Height", hue="Sex", kind="kde")
plt.ylabel("Frequency")
plt.title("Atheletes Height Histogram")
plt.show()
```

```python
sns.barplot(data=data, x="Medal", y="Height", hue="Sex")
plt.title("Medals by Height")
plt.show()
```

```python
sns.catplot(x='Medal', y='Height', hue='Sex', col='Season', data=data, kind='bar')
```
